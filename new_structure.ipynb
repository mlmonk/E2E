```python
base_dir = 'data/jurik_data/e2e_aligned'

data = pd.read_csv(base_dir+'/trainset_e2e [aligned].csv') 
ref = data.loc[:, 'ref']
align = data.loc[:, 'alignment']
mr = data.loc[:,'mr']
stop = ref.str.find('.')
align[1]
align = align.str.replace(')','')
align = align.str.replace('(',',')
align = align.str.strip(',')
split = align.str.split(',') 
mr = mr.str.split(',') 
```

```python
a = [0 for i in range(len(ref))]
b = [0 for i in range(len(ref))]

order = [0 for i in range(len(ref))]

len(ref)
for i in range(len(ref)):
    a[i] = [m.start() for m in re.finditer('\.', ref[i])]
for i in range(len(mr)):
    b[i] = mr[i]

```

```python
for i in range(len(ref)):
    for j in range(len(a[i])):
        for k in range(len(split[i])):
            if re.sub('\D','',split[i][k])=='':
                break
            if a[i][j] < int(re.sub('\D','',split[i][k])):
                split[i].insert(k,'period')
                j += j
        
for i in range(len(ref)):  
    if type(split[i])==float:
        print(i)
        continue
    order[i] = list(set(split[i]))
    order[i].sort(key=split[i].index)

order[2012]
```

```python
mr_attr_dict = []
for i in range(0,len(mr_test)):
    mr_attr_dict.insert(i, {})
    for attr_val in mr_test[i].split(','):
        attr_val = attr_val.replace('familyFriendly', 'familyfriendly')
        attr_val = attr_val.replace('customer rating', 'customerrating')
        attr_val = attr_val.replace('eatType', 'eattype')
        attr_val = attr_val.replace('priceRange', 'pricerange')
        #print ('Attr:', attr_val)
        attr_val = attr_val.strip()
        attr_key = attr_val[0:attr_val.find('[')]
        #print ('Key:', attr_key)
        mr_attr_dict[i][attr_key] = attr_val
        
    mr_attr_dict[i]['period'] = 'period[.]'
        
#mr_attr_dict
```
