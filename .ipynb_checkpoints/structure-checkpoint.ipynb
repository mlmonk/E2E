### Create a dictionary of attributes

```python

mr_attr_dict = []
for i in range(0,len(mr)):
    mr_attr_dict.insert(i, {})
    for attr_val in mr[i].split(','):
        #print ('Attr:', attr_val)
        attr_val = attr_val.strip()
        attr_key = attr_val[0:attr_val.find('[')]
        #print ('Key:', attr_key)
        mr_attr_dict[i][attr_key] = attr_val
```
